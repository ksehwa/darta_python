```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt

x = [1,4,9,16,25, 36, 49, 64]
plt.plot(x)
plt.show()
```

그래프 모양, 색 지정

```python
plt.plot(x, color='r')
plt.show(x)
```

```python
plt.plot(x, 'or')    # 모양과 색을 한번에 변경
plt.show(x)
```

```python
y = [i for i in range(1,9)]
plt.plot(x,y)           # x,y 축 그리기
plt.xlabel('x')         # x축 이름 설정
plt.ylabel('y')         # y축 이름 설정
plt.title('matplotlib sample')      # 그래프 이름 설정
plt.show()
```

df에서 나이가 40보다 많고, 점수가 89보다 높은 사람의 이름만 출력

```python
import pandas as pd
data = {'name' : ['Mark','Jane','Chris','Ryan'],
        'age' : [33,32,44,42],
        'score':[91.3, 83.4, 77.5, 87.7] } 
df = pd.DataFrame(data)

df.name[(df.age > 40) & (df.score > 80)]
```

점수 기준으로 내림차순 정렬

```python
df.sort_values(by = 'score', ascending=False)
```

df 에서 이름에 a가 들어간 사람들의 이름만 출력

```python
df[df.name.str.find('a') > -1]
```
